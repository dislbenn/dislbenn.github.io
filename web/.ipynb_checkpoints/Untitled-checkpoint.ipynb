```python
import json
path = "/Users/cerser/Desktop/pydata-book/datasets/bitly_usagov/example.txt"
records = [json.loads(line) for line in open(path)]
```

```python
print(type(records))
```

```python
print(records[0])
print(records[0]['tz'])
```

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
print(len(time_zones))

d = {}
for tz in time_zones:
    if tz in d:
        d[tz] += 1
    else:
        d[tz] = 1
        
for tz in d:
    print("%50s%10d\n" % (tz, d[tz]))
```

```python
from pandas import DataFrame, Series
import pandas as pd

frame = DataFrame(records)
print(frame['tz'][:10])
l = frame['tz'].value_counts()
print(l[:10])

clean_tz = frame['tz'].fillna('Missing')
clean_tz[clean_tz == ''] = 'Unknown'
l = clean_tz.value_counts()
```
